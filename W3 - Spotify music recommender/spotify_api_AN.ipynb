# Scraping songs uri, names and artists

```python
# lists with ids, song names and artists names

playlist_uri = []
playlist_song_names = []
playlist_artists_names = []

# begining at first page = index(0)
next_page_index = 0

# there are 10,000 songs in this list, so we have 100 pages x 100 songs each = 10000
for page in range(100):
    
    # make query to generate data
    spotify_playlist = sp.user_playlist_tracks('spotify}', '6FKDzNYZ8IW1pvYVF4zUN2', limit=100, offset=next_page_index)
    
    # there are 100 songs in each page, so we are going to iterate over each one and add data to our lists
    for index in range(100):
        
        # generating song's id, name and artist
                                
        spotify_playlist_uri_loc = spotify_playlist['items'][index]['track']['uri']
        spotify_playlist_song_name_loc = spotify_playlist['items'][index]['track']['name']
        spotify_playlist_artists_name_loc = spotify_playlist['items'][index]['track']['artists'][0]['name']
        
        # appending information to the lists
        playlist_uri.append(spotify_playlist_uri_loc)
        playlist_song_names.append(spotify_playlist_song_name_loc)
        playlist_artists_names.append(spotify_playlist_artists_name_loc)

    # going to the next page
    next_page_index += 100
    print(f"I'm on a {next_page_index} song")
    print()

    # timer in order to prevent being banned
    
    wait_time = randint(4, 7)
    print("I will sleep for " + str(wait_time) + " seconds.")
    print()
    sleep(wait_time)
```

# Generating list of dictionaries that containing audio features for each song 

```python
# generating list of all audio features

list_of_audio_features_spotify_playlist = []


starting_index = 0
ending_index = 100

for i in range(100):
    list_of_audio_features = sp.audio_features(playlist_uri[starting_index:ending_index])
    list_of_audio_features_spotify_playlist.extend(list_of_audio_features)
    
    starting_index += 100
    ending_index += 100
    
    
    
```

# Prepare dictionary for dataframe

```python
# prepeare a dictionary for data frame containing all audio features

features_spotify_playlist = {'danceability': [], 'energy': [], 'key': [], 'loudness': [], 'mode': [], 'speechiness': [],
            'acousticness': [], 'instrumentalness': [],
            'liveness': [], 'valence': [], 'tempo': [], 'type': [], 'id': [], 'uri': [], 'track_href': [],
            'analysis_url': [], 'duration_ms': [],
            'time_signature': []}


## fill the dictionary with data from audio features and be sure that all values are stored in list in order to generate later dataframe
## filling the dicionary with for loop, each position/index (song) represents dictionary with all audio features of that song

for song in list_of_audio_features_spotify_playlist:
    for key in features_spotify_playlist.keys():
        features_spotify_playlist[key].append(song[key])
        

```

```python
### generating dataframe that contains all audio features 

spotify_features_df = pd.DataFrame(features_spotify_playlist)


### preparing dictionary of songs titles and artists for DF


spotify_playlist_info_dictionary = {'Title':playlist_song_names,
                         'Artist':playlist_artists_names}


### generating dataframe that contais all song titles and artists names

spotify_playlist_songs_info_df = pd.DataFrame(spotify_playlist_info_dictionary)


### concatenating both dataframes top hot 100 songs and their audio features

spotify_playlist_df = pd.concat([spotify_playlist_songs_info_df, spotify_features_df], axis=1)
```
